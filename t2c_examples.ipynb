```python
# Function to pull datasets from SCXA by accession & save them locally if not already downloaded
import os
import requests

def get_scxa_h5ad(accession):
    file_name = f'h5ad_cache/{accession}_modified.project.h5ad'
    if not os.path.exists(file_name):
        url = f'https://ftp.ebi.ac.uk/pub/databases/spot/scxa/{accession}_modified.project.h5ad'
        with requests.get(url, stream=True) as r:
            r.raise_for_status()
            with open(file_name, 'wb') as f:
                for chunk in r.iter_content(chunk_size=8192):
                    f.write(chunk)
    return file_name
```
